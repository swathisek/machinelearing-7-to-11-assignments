```python
working Principle Of KNN

    The kNN algorithm is belongs to the family of instance-based, competitive learning and lazy learning algorithms.

Instance-based algorithms are those algorithms that model the problem using data instances (or rows) in order to make predictive decisions. The kNN algorithm is an extreme form of instance-based methods because all training observations are retained as part of the model.
 
    It is a competitive learning algorithm, because it internally uses competition between model elements (data instances) in order to make a predictive decision. The objective similarity measure between data instances causes each data instance to compete to “win” or be most similar to a given unseen data instance and contribute to a prediction.

Lazy learning refers to the fact that the algorithm does not build a model until the time that a prediction is required. It is lazy because it only does work at the last second. This has the benefit of only including data relevant to the unseen data, called a localized model. A disadvantage is that it can be computationally expensive to repeat the same or similar searches over larger training datasets.
 
kNN is powerful because it does not assume anything about the data, other than a distance measure can be calculated consistently between any two instances. As such, it is called non-parametric or non-linear as it does not assume a functional form.

Uses Euclidean distance to measure the distance between the points

Euclidean distance is the most commonly used distance measure.
Euclidean distance also called as simply distance. 
The usage of Euclidean distance measure is highly recommended when data is dense or continuous. 
Euclidean distance is the best proximity measure. 
The Euclidean distance between two points is the length of the path connecting them.
The Pythagorean theorem gives this distance between two points. 


```

```python
# The columns that we will be making predictions with.
xcolumn = [ 'G', 'GS', 'MP', 'FG', 'FGA', 'FG.', 'x3P', 'x3PA',
 'x3P.', 'x2P', 'x2PA', 'x2P.', 'eFG.', 'FT', 'FTA', 'FT.', 'ORB', 'DRB', 'TRB', 'AST',
 'STL', 'BLK', 'TOV', 'PF']
```

```python
#Selecting coloumns 
xcolumn=nba.iloc[: , 4:28]
xcolumn.head(10)
```

```python
import numpy as np
from sklearn.preprocessing import Imputer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import accuracy_score
```

```python
#For treating the NAN and the Null values, Imputer, is used  to replace the NAN values with the median
imp = Imputer(missing_values="NaN", strategy='median', axis=0)

xcolumn = imp.fit_transform(xcolumn)
xcolumn.shape
```

```python

# The column that we want to predict.
ycolumn = nba.iloc[: ,28]
ycolumn=np.array(ycolumn)
ycolumn.shape
```

```python
# Create the knn model assuming the number of neighbours as 5
knn = KNeighborsRegressor(n_neighbors=5)
```

```python
#Splitting the dataset as training and testing sets,allocating 30 percentage of data  for testing and the rest for training
x_train, x_test, y_train, y_test = train_test_split(
 xcolumn, ycolumn, test_size = 0.3, random_state = 100)
```

```python
# Fit the model on the training data.
knn.fit(x_train,y_train)
```

```python
# Make predictions on the test set using the fit model.
ypredict= knn.predict(x_test)
ypredict


```

```python
#Since Regressor returns the exact value this is the score fetched by the nba_Players for the given year
result=knn.score(x_test,y_test)
np.round(result,3)
```
